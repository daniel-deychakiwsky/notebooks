```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cross_entropy(y_true, y_hat):
    return -1.0 * ((y_true * np.log(y_hat)) + ((1.0 - y_true) * np.log(1.0 - y_hat)))
```

```python
prior = 0.3
dataset_size = np.arange(10, 50000, 100)
monte_carlo = 1000

r = []

for ds_size in dataset_size:
    y_true = np.random.binomial(1, prior, (ds_size, monte_carlo))
    mean_cross_entropy = cross_entropy(y_true, prior).mean(axis=1)
    r.append(mean_cross_entropy.mean())
```

```python
plt.plot(dataset_size, r, label='mean_cross_entropy');
plt.axhline(cross_entropy(prior, prior), label='prior_entropy', color='orange');
plt.xlabel('training_data_size');
plt.title('"avg log-loss per impression if the model predicted prior CTR"');
plt.legend();
```
